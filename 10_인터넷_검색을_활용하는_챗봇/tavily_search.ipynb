```python
import os
from dotenv import load_dotenv
load_dotenv(dotenv_path="../.env")
```

```python
from langchain_community.tools import TavilySearchResults

tavily_search = TavilySearchResults(
    max_results=5
)
```

```python
# 질문하기
res = tavily_search.invoke({"query": "갤럭시 Z 폴드7 에 대한 반응은?"})
for r in res:
    print(r)
```

```python
# 주피터 노트북에서 가독성 좋게 출력하기
from IPython.display import JSON
JSON(res).data
```

```python
# 웹 페이지 전문 불러오기
from tavily import TavilyClient

client = TavilyClient()

content = client.search(
    "갤럭시 Z 폴드7 에 대한 반응은?",  
    search_depth="advanced",
    include_raw_content=True,
)

JSON(content).data
```

#### 인터넷에서 자료 조사 후 기사 쓰는 기자 만들기

```python
# 질문에 대한 자료 검색
from langchain.adapters.openai import convert_openai_messages
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")

# Step 1. 질문에 대한 자료를 TavilyClient로 가져오기
query = "갤럭시 Z 폴드7 에 대한 반응은?"

content = client.search(
    query, 
    include_raw_content=True,
    search_depth="advanced"
)["results"]

content
```

```python
# 프롬프트 설정
prompt = [{
    "role": "system",
    "content":  (
        "당신은 신문기사를 쓰는 기자 AI입니다.  \n"
        "당신은 주어진 정보를 바탕으로 객관적이고 체계적으로 작성된 기사를 써야 합니다. \n"
    )
}, {
    "role": "user",
    "content": (
        f'정보: """{content}"""\n\n' 
        f'위의 정보를 사용하여, 다음 질문에 대해 자세한 보고서를 한국어로 작성하세요: "{query}" \n'
        '—신문기사 형식을 사용하되, MLA를 준수하는 markdown 문법을 사용해주세요.'
        '—활용한 자료는 출처를 명시하세요.'
    )
    
}]
```

```python
# 오픈AI 를 랭체인으로 실행하고 출력하기

report = llm.invoke(prompt).content
print(report)
```
