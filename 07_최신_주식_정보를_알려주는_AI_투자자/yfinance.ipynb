```python
import yfinance as yf

# Microsoft(MSFT)에 대한 Ticker 객체 생성
msft = yf.Ticker("MSFT")

# Ticker 객체의 정보 출력
display(msft.info)
```

```python
# 최근 5일간의 주가 정보 확인하기
hist = msft.history(period="5d")
display(hist)
```

```python
# 주식 종목의 추천 여부 보기
msft.recommendations
```
